```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbr
```

```python
df = pd.read_csv("2019.csv")
df
```

```python
dataIndex = ['Mean','Median','Modus','Range','Q1','Q3','IQR','Standar Deviasi','Variance']
dataColumns = ['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices','Generosity','Perceptions of corruption']
s = [0,1,2,3,4,5,6]
df2 = pd.DataFrame(index = dataIndex,columns = dataColumns)

for i in range(len(dataColumns)):
    df2[dataColumns[i]] = [df[dataColumns[i]].mean(),df[dataColumns[i]].median(),df[dataColumns[i]].mode(),(df[dataColumns[i]].max())-(df[dataColumns[i]].min()),np.quantile(df[dataColumns[i]], .25),np.quantile(df[dataColumns[i]], .75),(np.quantile(df[dataColumns[i]], .75))-(np.quantile(df[dataColumns[i]], .25)),np.std(df[dataColumns[i]]),np.var(df[dataColumns[i]])]
df2
```

### Z-Score

```python
dfzscore = df.drop(["Overall rank"], axis=1)
for i in dataColumns :
    zscore = i + "z-score"
    dfzscore[zscore] = (dfzscore[i] - dfzscore[i].mean())/dfzscore[i].std(ddof=1)
dfzscore
```

### High Ouliers & Low Outliers

```python
# Menentukan High Ouliers & Low Outliers
dataIndex = ['ZScore-High','ZScore-Low','IQR-High','IQR-Low']
dataColumns = ['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices','Generosity','Perceptions of corruption']
s = [0,1,2,3,4,5,6]
zdf = pd.DataFrame(index = dataIndex,columns = dataColumns)

for i in range(len(dataColumns)):
    zdf[dataColumns[i]] = [df[dataColumns[i]].mean() + (2*np.std(df[dataColumns[i]])),df[dataColumns[i]].mean() - (2*np.std(df[dataColumns[i]])), np.quantile(df[dataColumns[i]], .75)+1.5 * (np.quantile(df[dataColumns[i]], .75))-(np.quantile(df[dataColumns[i]], .25)), np.quantile(df[dataColumns[i]], .25)-1.5 * (np.quantile(df[dataColumns[i]], .75))-(np.quantile(df[dataColumns[i]], .25))]
zdf
```

### Boxplot

```python
# BoxPlot
x = []
for i in range(len(dataColumns)):
    x.append(df[dataColumns[i]])
fig, axis = plt.subplots(figsize=(16,8))
axis.boxplot(x, vert=True)
plt.show()
```
